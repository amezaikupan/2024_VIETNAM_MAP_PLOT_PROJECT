# Using GeoJson to plot Vietnam map

Data source: [GADM database](https://gadm.org/download_country.html#google_vignette)

This is a tutorial on how to use GeoJson files to plot Vietnam specific map plot.

## 1. Setting things up

There are a few things we have to set up before plotting our maps.

### 1.1 Importing necessary libraries

```python
import geopandas as gpd

import geoplot
import geoplot.crs as gcrs

import random
```

### 1.2 Load our GeoJSON files

There are 3 files that all depicts Vietnam maps but in increasing complexity level.
- Level 1: Provinces/administrative city level.
- Level 2: Districs level.
- Level 3: Village level.

```python
data_lv1 = gpd.read_file('gadm41_VNM_1.json')
data_lv2 = gpd.read_file('gadm41_VNM_2.json')
data_lv3 = gpd.read_file('gadm41_VNM_3.json')
```

### 1.3 Other ultilities

We will need to do some cleaning with the dataset as well. One of the task is to convert Vietnamese name to English, which essentially means removing all the diacritics used.

Code: [J2TEAM on Github](https://gist.github.com/J2TEAM/9992744f15187ba51d46aecab21fd469)

```python
s1 = u'ÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚÝàáâãèéêìíòóôõùúýĂăĐđĨĩŨũƠơƯưẠạẢảẤấẦầẨẩẪẫẬậẮắẰằẲẳẴẵẶặẸẹẺẻẼẽẾếỀềỂểỄễỆệỈỉỊịỌọỎỏỐốỒồỔổỖỗỘộỚớỜờỞởỠỡỢợỤụỦủỨứỪừỬửỮữỰựỲỳỴỵỶỷỸỹ'
s0 = u'AAAAEEEIIOOOOUUYaaaaeeeiioooouuyAaDdIiUuOoUuAaAaAaAaAaAaAaAaAaAaAaAaEeEeEeEeEeEeEeEeIiIiOoOoOoOoOoOoOoOoOoOoOoOoUuUuUuUuUuUuUuYyYyYyYy'
def remove_accents(input_str):
	s = ''
	for c in input_str:
		if c in s1:
			s += s0[s1.index(c)]
		else:
			s += c
	return s
```

## 2. Understand and processing our data

Let's see how our data is structured.

```python
# Level 1 complexity
print(data_lv1.columns)
data_lv1.head(3)
```

Columns that we will use:
- NAME_1: Name of province in Vietnamese.
- VARNAME_1: Name of privince in English.
- TYPE_1: Administrative level in Vietnamese.
- ENGTYPE_1: Administrative level in English.

---

For level 2 and 3:

```python
# Level 2 complexity
print(data_lv2.columns)
data_lv2.head(3)
```

```python
# Level 3 complexity
print(data_lv3.columns)
data_lv3.head(3)
```

Noticed that province name of level 2 and level 3, along with district name in level 3 doesn't have corresponding English name. Let's transform the Vietnamese name into English columns.

```python
data_lv2['NL_NAME_1'] = data_lv2['NAME_1'].transform(lambda x: remove_accents(x))
data_lv3['NL_NAME_1'] = data_lv3['NAME_1'].transform(lambda x: remove_accents(x))
data_lv3['NL_NAME_2'] = data_lv3['NAME_2'].transform(lambda x: remove_accents(x))
```

Overall, these are the columns we will work on in level 2 and 3:
- NAME_1
- NL_NAME_1
- NAME_2
- VARNAME_2
- NL_NAME_2

With level 3:
- NAME_3
- VARNAME_3

## 3. Plotting the maps

### 3.1 Ploting at country level

With level 1 complexity:

```python
# Using gepplot.polyplot produce map with quite low resolution
geoplot.polyplot(data_lv1, projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```

With level 2 complexity: 

```python
# Using gepplot.polyplot produce map with quite low resolution
geoplot.polyplot(data_lv2, projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```

With level 3 complexity:

```python
# Using gepplot.polyplot produce map with quite low resolution
geoplot.polyplot(data_lv3, projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```

### 3.2 Ploting at provincial level

Let's choose a province to plot

```python
province = 'TienGiang'
```

Using level 1 complexity:

```python
geoplot.polyplot(data_lv1[data_lv1['VARNAME_1'] == province], projection=gcrs.AlbersEqualArea(), edgecolor='black', facecolor='pink', linewidth=.3,
    figsize=(12, 8))
```

```python
geoplot.polyplot(data_lv2[data_lv2['NL_NAME_1'] == province], projection=gcrs.AlbersEqualArea(), edgecolor='black', facecolor='pink', linewidth=.3,
    figsize=(12, 8))
```

```python
geoplot.polyplot(data_lv3[data_lv3['NL_NAME_1'] == province], projection=gcrs.AlbersEqualArea(), edgecolor='black', facecolor='pink', linewidth=.3,
    figsize=(12, 8))
```

### 3.3 Plotting at district level

We can only plot with district level with data comlexity of level 2 and 3.

```python
# district  = 'your choice'
# Generate random district
district = random.choice(data_lv2[data_lv2['NL_NAME_1'] == province]['VARNAME_2'].unique())
```

With level 2:

```python
geoplot.polyplot(data_lv2[(data_lv2['NL_NAME_1'] == province) & (data_lv2['VARNAME_2'] == district)], projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```

With level 3:

```python
geoplot.polyplot(data_lv3[(data_lv3['NL_NAME_1'] == province) & (data_lv3['NL_NAME_2'] == district)], projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```

### 3.4 Plotting at ward/village level

We can only plot this with complexity of level 3.

```python
# ward  = 'your choice'
# Generate random ward
ward = random.choice(data_lv3[(data_lv3['NL_NAME_1'] == province) & (data_lv3['NL_NAME_2'] == district)]['VARNAME_3'].unique())
```

```python
geoplot.polyplot(data_lv3[(data_lv3['NL_NAME_1'] == province) & (data_lv3['NL_NAME_2'] == district) & (data_lv3['VARNAME_3'] == ward)], projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
```
